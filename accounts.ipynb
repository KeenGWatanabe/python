```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the TB.csv file into a Pandas DataFrame , usecols=['REVENUE', 'NETT']
df = pd.read_csv('TB.csv', index_col='description')
# Plot the NETT and REVENUE columns
tb = df.T
# Convert the NETT and REVENUE columns to numeric data types
tb['NETT'] = tb['NETT'].astype('float')
tb['REVENUE'] = tb['REVENUE'].astype('float')

tb.plot(x='NETT', y='REVENUE', kind='line')

plt.show()
```
