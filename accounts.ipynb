```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#this dataset is re-input from csv file as reference  
data = {'account': ['ASSETS', 'LIABILITIES', 'REVENUE', 'COST OF SALES', 'MARKETING','GENERAL EXPENSES','NETT'],
        'JAN': [0, 7014, 39127, 35467, 22487, 3321, 22148.83],
        'FEB': [0, 2581.0, 34234, 19321, 14092, 3304, 2483.36],
       }  
df = pd.DataFrame(data)

#melt down dataframe for plotting
tb = pd.melt(df, id_vars=["account"], value_vars=["JAN", "FEB"]) \
    .rename({"variable": "month"}, axis=1)
#plotting the graph with line
tb[tb["account"] == "REVENUE"].plot(x='month', y='value', kind='line')
plt.show()
```
